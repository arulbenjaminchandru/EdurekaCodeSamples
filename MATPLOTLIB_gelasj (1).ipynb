```python
x = np.arange(1,2,0.1)
```

```python
x*=10
```

```python
plt.plot(x,[a*100 for a in x])
plt.show()
```

```python
plt.plot(x,[a**2 for a in x])
plt.plot(x,[a**3 for a in x])
plt.plot(x,[a**4 for a in x])
#plt.plot(x,[a*10 for a in x])
plt.show()
```

```python
plt.plot(x,[a**2 for a in x],x,[a**3 for a in x],x,[a**4 for a in x])
plt.grid()
plt.axis([-10,10,-1000,5000])
plt.show()
```

```python
plt.plot(x,[a+2 for a in x],x,[a+3 for a in x],x,[a+4 for a in x],x,[a+10 for a in x])
plt.grid()
#plt.xlim([9,20])
plt.xlabel('This is my x-axis')
plt.ylabel('This is my y-axis')
plt.title('This is the TITLE of my plot')
plt.legend()
#plt.ylim([0,5000])
plt.show()
```

```python
plt.plot(x,[a+2 for a in x],'b',linestyle='-',marker='*',label='Multiple of 2')
plt.plot(x,[a+3 for a in x], 'c',linestyle='--',marker='o',label = 'Mul of 3')
plt.plot(x,[a+4 for a in x], 'r',linestyle='-.',marker='D',label = 'Mul of 4')
plt.grid()
plt.xlabel('This is my x-axis')
plt.ylabel('This is my y-axis')
plt.title('This is the TITLE of my plot')
plt.legend()
#plt.plot(x,[a*10 for a in x])
plt.savefig('Plot saved live in class 901.png')
plt.show()
```
